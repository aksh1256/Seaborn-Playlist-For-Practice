```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = sns.load_dataset("tips")
data
```

```python
fg = sns.FacetGrid(data, col="sex", hue="day")
fg.map(plt.scatter, "total_bill", "tip").add_legend()
plt.show()
```

```python
fg = sns.FacetGrid(data, col="day", hue="sex")
fg.map(plt.bar, "total_bill", "tip").add_legend()
plt.show()
```

```python
fg = sns.FacetGrid(data, col="sex", hue="day", palette="summer")
fg.map(plt.bar, "total_bill", "tip", edgecolor="r").add_legend()
plt.show()
```
