```python
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
var = np.linspace(1,10,20).reshape(4,5)
var
# Heatmap plot required only one clolumn for graph a plot
# here we create a 2- dimensional array with the help of pandas linspace function
```

```python
sns.heatmap(var, vmin=0, vmax=12)
plt.show()
```

```python
sns.heatmap(var, vmin=0, vmax=12, cmap="PRGn")
plt.show()
```

```python
y = {"fontsize":10, "color":"r"}
sns.heatmap(var, vmin=0, vmax=10, cmap="PRGn", annot=True, annot_kws=y)
plt.show()
```

```python
z = {"fontsize":12, "color":"b"}
sns.heatmap(var, vmin=0, vmax=10, cmap="PRGn", annot=True, annot_kws=z, linewidth=4)
plt.show()
```

```python
z = {"fontsize":13, "color":"orange"}
sns.heatmap(var, vmin=0, vmax=10, cmap="PRGn", annot=True, annot_kws=z, linewidth=10, linecolor="y")
plt.show()
```

```python
z = {"fontsize":13, "color":"orange"}
sns.heatmap(var, vmin=0, vmax=10, cmap="PRGn", annot=True, annot_kws=z, linewidth=10, linecolor="y", cbar=False)
plt.show()
```

```python
z = {"fontsize":13, "color":"orange"}
sns.heatmap(var, vmin=0, vmax=10, cmap="PRGn", annot=True, annot_kws=z, linewidth=10, linecolor="y", cbar=False, xticklabels=False)
plt.show()
```

```python
z = {"fontsize":13, "color":"orange"}
sns.heatmap(var, vmin=0, vmax=10, cmap="PRGn", annot=True, annot_kws=z, linewidth=10, linecolor="y", cbar=False, xticklabels=False, yticklabels=False)
plt.show()
```

```python
z = {"fontsize":13, "color":"orange"}
v = sns.heatmap(var, vmin=0, vmax=10, cmap="PRGn", annot=True, annot_kws=z, linewidth=10, linecolor="y", cbar=False, xticklabels=False, yticklabels=False)
# v.set(xlabel="python", ylabel="java")
plt.show()
```

```python
z = {"fontsize":13, "color":"orange"}
v = sns.heatmap(var, vmin=0, vmax=10, cmap="PRGn", annot=True, annot_kws=z, linewidth=10, linecolor="y", cbar=False, xticklabels=False, yticklabels=False)
v.set(xlabel="python", ylabel="java")
plt.show()
```

```python
####
var1 = np.linspace(1,10,10).reshape(2,5)
var1
```

```python
sns.heatmap(var1, vmin=0, vmax=10, cmap="PuOr", annot=True)
plt.show()
```

```python
ar = np.array([["a0","a1","a2","a3","a4"],
              ["b0","b1","b2","b3","b4"]])
ar
```

```python
sns.heatmap(var1, vmin=0, vmax=10, cmap="PuOr", annot=ar, fmt="s")
plt.show()
```
