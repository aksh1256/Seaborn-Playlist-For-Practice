```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.catplot(data=var)
plt.show()
```

```python
sns.catplot(x="size", y="tip", data=var, hue="sex")
plt.show()
```

```python
sns.catplot(x="size", y="tip", data=var, hue="sex", palette="hsv", size=6)
plt.show()
```
