**Styling Plot In Seaborn**

```python
# Seaborn Feagure Style
# Removing Axes Spines
# Scale And Context
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = sns.load_dataset("tips")
data
```

```python
sns.boxplot(y=data["total_bill"])
plt.show()
```

```python
# Barplot
sns.set_style("darkgrid")    #White, dark, whitegrid, darkgrid, ticks
sns.barplot(x="day", y="total_bill", data=data)
plt.show()
```

```python
# Violin Plot
sns.set_style("white")
sns.violinplot(x="day", y="total_bill", data=data, hue="sex")
sns.despine() # Remove two side lines
plt.show()
```

```python
# Strip Plot
sns.set_style("white")
plt.figure(figsize=(10,3))     #Using this we change the size
sns.stripplot(x="day", y="total_bill", data=data)
sns.despine()
plt.show()
```

```python
# Bar Plot
sns.set_style("white")
sns.set_context("poster",font_scale=0.4)    # paper, notebook, talk, poster
sns.barplot(x="day", y="total_bill", data=data, palette="cool")
plt.show()
```
