```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = sns.load_dataset("tips")
data
```

```python
sns.pairplot(data)
```

```python
sns.pairplot(data, hue="sex")
plt.show()
```

```python
sns.pairplot(data, vars=["total_bill", "tip"], hue="sex", hue_order=["Female", "Male"])
plt.show()
```

```python
sns.pairplot(data, vars=["total_bill", "tip"], hue="sex", palette="hot")
plt.show()
```

```python
sns.pairplot(data, hue="sex", palette="hot", x_vars=["total_bill", "tip"])
plt.show()
```

```python
sns.pairplot(data, hue="sex", palette="hot", y_vars=["tip", "total_bill"])
plt.show()
```

```python
sns.pairplot(data, hue="sex", kind="reg")      # kind=["scatter", "kde", "hist", "reg"]
plt.show()
```

```python
sns.pairplot(data, hue="sex", kind="kde", diag_kind="hist")
plt.show()
```

```python
sns.pairplot(data, hue="sex", markers=["*", "<"])
plt.show()
```
