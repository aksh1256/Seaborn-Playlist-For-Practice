**Violin Plot**

```python
# A Violin Plot plays a same role as a Box plot and whisker plot. It shows a distribution of quantitative data across several levels
#  of one (or more) categorical variables such that those distributions can be compared.

# import seaborn

# seaborn.set(style='whitegrid')
# data = seaborn.load_dataset("")
# seaborn.violinplot(x=" ", y=" ", data=data)
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
```

```python
data = sns.load_dataset("tips")
data
```

```python
sns.violinplot(x="day", y="total_bill", data=data)
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=data, hue="time")
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=data, linewidth=5)
```

```python
sns.violinplot(x="day", y="total_bill", data=data, linewidth=5, palette="cool")
plt.show()
```

```python
sns.violinplot(x="time", y="total_bill", data=data, order=["Dinner","Lunch"])
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=data, palette="cool", saturation=0.5)
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=data, color="g", linewidth=4)
```

```python
sns.violinplot(x="day", y="total_bill", data=data, hue="sex", split=True)
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=data, hue="sex", split=True, scale="width")  #scale="count","area","width"
plt.show()
```

```python
sns.violinplot(x="total_bill", y="day", data=data,)
# for horizontally
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=data, inner= "quart")   # inner=("quart", "quartile", "point", "stick", None)
```

```python
sns.violinplot(data["total_bill"])
```
