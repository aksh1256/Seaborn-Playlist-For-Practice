```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.set(style="whitegrid")
```

```python
sns.boxplot(x="day", y="total_bill", data=var)
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", data=var, hue="time")
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", data=var, hue="sex", color="r")
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", data=var, hue="sex", color="g", order=["Sat","Thur","Sun","Fri"])
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", data=var, hue="sex", color="r", showmeans=True)
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", data=var, color="r", showmeans=True)
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", data=var, color="w", showmeans=True, meanprops={"marker":"+", "markeredgecolor":"b"})
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", data=var, showmeans=True, linewidth=6)
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", data=var, showmeans=True, linewidth=6, palette="plasma")
plt.show()
```

```python
sns.boxplot(x="total_bill", y="day", data=var, linewidth=6, palette="plasma")
plt.show()
```

```python
sns.boxplot(data=var, palette="plasma", linewidth=3, orient="h")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(y=var["total_bill"])
plt.show()
```
