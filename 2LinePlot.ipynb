```python
import matplotlib.pyplot as plt

var1 = [1, 3, 5, 7, 9]
var2 = [2, 4, 3, 4, 1]

plt.plot(var1, var2)
# plt.show()
```

```python
import seaborn as sns
import pandas as pd

var1 = [1, 3, 5, 7, 9]
var2 = [2, 4, 3, 4, 1]
df = pd.DataFrame({"var1":var1, "var2": var2})

sns.lineplot(x="var1", y="var2", data=df)
# plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
y = sns.load_dataset("penguins")
y
```

```python
sns.lineplot(x="species",y="island",data=y)
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=y)
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=y,hue="sex")
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=y,hue="sex",style="sex")
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=y,hue="sex",style="sex",palette="Accent")
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=y,hue="sex",style="sex",palette="Accent",markers=["o",">"])
plt.show()
```
