**Strip Plot**

```python
# Strip plot is very simple to understand. It is basically a scatter plot that differentiates different categories.
# Strip plots are considered a good alternative to a box plot or a violin plot.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.stripplot(x="day", y="total_bill", data=var, palette="hot")
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=var, palette="hsv", hue="sex")
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=var, hue="sex", palette="hsv", linewidth=2)
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=var, hue="sex", palette="hsv", linewidth=1, edgecolor="r")
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=var, hue="sex", palette="hsv", linewidth=1, edgecolor="r", jitter=0.2)
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=var, hue="sex", palette="hsv", linewidth=2, edgecolor="r", size=10)
plt.show()
```

```python
# m = {"Male":"*", "Female":"o"}
sns.stripplot(x="day", y="total_bill", data=var, hue="sex", palette="hsv", linewidth=1, edgecolor="r", size=10, marker="*")
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=var, hue="sex", linewidth=2, edgecolor="r", alpha=0.5)
plt.show()
```

```python
sns.stripplot(x=var["total_bill"])
plt.show()
```

```python
sns.stripplot(y=var["total_bill"],palette="brg")
plt.show()
```
