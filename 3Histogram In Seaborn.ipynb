```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
var = sns.load_dataset("penguins")
var
```

```python
sns.displot(var["flipper_length_mm"])
plt.show()
```

```python
sns.displot(var["flipper_length_mm"], bins=[170,180,190,200,210,220,230,240])
plt.show()
```

```python
sns.displot(var["flipper_length_mm"], bins=[170,180,190,200,210,220,230,240], kde=True)
plt.show()
```

```python
sns.displot(var["flipper_length_mm"], bins=[170,180,190,200,210,220,230,240], kde=True, rug=True)
plt.show()
```

```python
sns.displot(var["flipper_length_mm"], bins=[170,180,190,200,210,220,230,240], kde=True, rug=True, color="Orange")
plt.show()
```

```python
sns.displot(var["flipper_length_mm"], bins=[170,180,190,200,210,220,230,240], kde=True, rug=True, color="Black", log_scale=True)
plt.show()
```

```python
# log_scale without bins[]
sns.displot(var["flipper_length_mm"], kde=True, rug=True, color="Blue", log_scale=True)
plt.show()
```
