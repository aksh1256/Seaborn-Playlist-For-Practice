```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = sns.load_dataset("penguins")
var
```

```python
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=var)
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=var, hue="sex")
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=var, hue="sex", style="sex")
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=var, hue="sex", style="sex", size="sex")
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=var, hue="sex", style="sex", size="sex", sizes= (30,30))
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=var, hue="sex", style="sex", size="sex", sizes=(60, 60), palette="brg")
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=var, hue="sex", style="sex", size="sex", sizes=(60,60), palette="gist_ncar", alpha=0.5)
plt.show()
```

```python
m ={"Male":"*","Female":"o"}
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=var, hue="sex", style="sex", size="sex", sizes=(100,100), palette="gist_ncar", alpha=1, markers=m )
plt.show()
```
