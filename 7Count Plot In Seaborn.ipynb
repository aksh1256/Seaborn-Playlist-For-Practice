```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = sns.load_dataset("tips")
data
```

```python
sns.countplot(x="sex", data=data,)
plt.show()
```

```python
sns.barplot(x="sex", y="size", data=data)
plt.show()
```

```python
sns.countplot(x="sex", data=data,)
plt.show()
```

```python
sns.countplot(x="sex", data=data, hue="smoker")
plt.show()
```

```python
sns.countplot(x="sex", data=data, hue="smoker", palette="bwr")
```

```python
sns.countplot(x="sex", data=data, hue="smoker", color="r")
plt.show()
```

```python
sns.countplot(x="sex", data=data, hue="smoker", color="r", saturation=0.1)
plt.show()
# Saturation come upto 0-1
```
