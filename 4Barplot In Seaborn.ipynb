```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = sns.load_dataset("penguins")
var
```

```python
# With using dataset name .var
# sns.barplot(x=var.island, y=var.bill_length_mm)
# plt.show()

# Without using dataset name .var
sns.barplot(x="island", y="bill_length_mm", data=var)
plt.show()
```

```python
# We use hue for compare male and female in island hue="sex"
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex")
plt.show()
```

```python
order = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex", order=order)
plt.show()
```

```python
order1= ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex", order=order1, hue_order=["Female", "Male"])
plt.show()
```

```python
order1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex", order= order1, hue_order=["Female", "Male"], ci=100)
plt.show()
```

```python
order1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex", order= order1, hue_order=["Female", "Male"], n_boot=2)
plt.show()
```

```python
order1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex", order= order1, hue_order=["Female", "Male"], orient="v")
plt.show()
```

```python
order1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data=var, orient="v")
plt.show()
```

```python
order1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x="bill_depth_mm", y="bill_length_mm", data=var, orient="h")
plt.show()
```

```python
order1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex", color="Yellow", order=order1)
plt.show()
```

```python
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex", palette="icefire")
plt.show()
```

```python
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex", saturation=1)
plt.show()
```

```python
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex", saturation=1, errcolor="yellow")
plt.show()
```

```python
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex", saturation=1, errcolor="yellow", errwidth=10)
plt.show()
```

```python
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex", saturation=1, errcolor="yellow", errwidth=10, capsize=0.3)
plt.show()
```

```python
order1= ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex", order=order1, color="Pink", dodge=True)
plt.show()
```

```python
order1= ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex", saturation=100, order=order1, dodge=True, alpha=0.3)
plt.show()
```

```python
sns.set(style="darkgrid")
order1= ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex", saturation=100, order=order1, dodge=True, alpha=0.3)
plt.show()
```
